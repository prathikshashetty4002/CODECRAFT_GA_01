```python
from transformers import pipeline, set_seed
```

```python
generator = pipeline('text-generation', model='gpt2')
set_seed(42)
```

```python
# Example 1
output = generator("Once upon a time in India,", max_length=50, num_return_sequences=1)
print(output[0]['generated_text'])

# Example 2
output = generator("Artificial Intelligence will", max_length=50, num_return_sequences=1)
print(output[0]['generated_text'])

# Example 3
output = generator("The future of education is", max_length=60, num_return_sequences=1)
print(output[0]['generated_text'])
```
